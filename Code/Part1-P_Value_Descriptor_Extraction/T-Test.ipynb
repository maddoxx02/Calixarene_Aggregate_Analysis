## Filtering

### 1. Remove fields with ZEROS accross all Compounds 

```python
dump = []
for i in range(len(df_desc.columns)):
    if df_desc.iloc[:,i][0] == 0 and df_desc.iloc[:,i][1] == 0 and df_desc.iloc[:,i][2] == 0 and df_desc.iloc[:,i][3] == [0]:
        dump.append(i)




print('Removed ',len(dump),' columns with NULL (ZERO) Values')
```

```python
df_desc_filtered = df_desc.drop(df_desc.columns[dump], axis = 1).copy()
```

```python
print('Descriptors remaining = ',len(df_desc_filtered.columns))
```

```python
L1 = []
L2 = []
PV = []
for i in range(1,len(df_desc_filtered.columns)):

    for j in range(0,len(df_synthesis.columns)):

        res = ttest_ind(df_desc_filtered.iloc[:,i], df_synthesis.iloc[:,j])
        print('The P-value is ', res.pvalue)
        if res.pvalue <= 0.05:
            print('The columns ', df_desc_filtered.columns[i], ' and ', df_synthesis.columns[j], ' are statistically significant')
            L1.append(df_desc_filtered.columns[i])
            L2.append(df_synthesis.columns[j])
            PV.append(res.pvalue)
        
            
```

```python
p_val_df = pd.DataFrame([L1,L2,PV]).transpose()
p_val_df.columns = ['RD_KIT_Desc','Synthesis_Metric','P_Value']
```

```python
p_val_df
```

```python
p_val_df.sort_values(by=['P_Value'], ascending=False).reset_index(drop = True).to_csv('P_value_descs.csv')
```

```python
p_val_df_filtered = p_val_df.sort_values(by=['P_Value'], ascending=False).reset_index(drop = True).copy()
```
